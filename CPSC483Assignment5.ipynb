### 7.Using the second half of Python code for Figure 9.2 - Somple Gaussian Naive Bayes Classification from Statistics, Data Mining and Machine Learning in Astronomy, 2nd edition as a guide, plot the decision boundaries for each classifier and dataset. ### 

```
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import colors
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
#Bayes classifier dataset 1

features = ['x1', 'x2']
X = dataset1[features]
X = X.values
y = dataset1.label

gnb = GaussianNB() 
gnb.fit(X, y)

xlim = (-5, 42)
ylim = (-5, 42)

xx, yy = np.meshgrid(np.linspace(xlim[0], xlim[1], 71), np.linspace(ylim[0], ylim[1], 81))

Z = gnb.predict_proba(np.c_[xx.ravel(), yy.ravel()])

Z = Z[:, 1].reshape(xx.shape)

fig = plt.figure(figsize= (5, 3.75))
ax = fig.add_subplot(111)

ax.scatter(X[:, 0], X[:, 1], c=y, cmap = 'RdBu')

ax.contour(xx, yy, Z, [0.5], colors ='k')

ax.set_xlim(xlim)
ax.set_ylim(ylim)

print('Bayes Classifier w/ dataset 1:')
plt.show()

#Bayes classifier dataset 2

features = ['x1', 'x2']
X = dataset2[features]
X = X.values
y = dataset2.label

gnb = GaussianNB() 
gnb.fit(X, y)

xlim = (-5, 42)
ylim = (-5, 42)

xx, yy = np.meshgrid(np.linspace(xlim[0], xlim[1], 71), np.linspace(ylim[0], ylim[1], 81))

Z = gnb.predict_proba(np.c_[xx.ravel(), yy.ravel()])

Z = Z[:, 1].reshape(xx.shape)

fig = plt.figure(figsize= (5, 3.75))
ax = fig.add_subplot(111)

ax.scatter(X[:, 0], X[:, 1], c=y, cmap = 'RdBu')

ax.contour(xx, yy, Z, [0.5], colors ='k')

ax.set_xlim(xlim)
ax.set_ylim(ylim)

print('Bayes Classifier w/ dataset 2:')
plt.show()

#KNN classifier dataset 1

features = ['x1', 'x2']
X = dataset1[features]
X = X.values
y = dataset1.label

kneighs = KNeighborsClassifier(n_neighbors= 41) 
kneighs.fit(X, y)

xlim = (-5, 42)
ylim = (-5, 42)

xx, yy = np.meshgrid(np.linspace(xlim[0], xlim[1], 71), np.linspace(ylim[0], ylim[1], 81))

Z = kneighs.predict_proba(np.c_[xx.ravel(), yy.ravel()])

Z = Z[:, 1].reshape(xx.shape)

fig = plt.figure(figsize= (5, 3.75))
ax = fig.add_subplot(111)

ax.scatter(X[:, 0], X[:, 1], c=y, cmap = 'RdBu')

ax.contour(xx, yy, Z, [0.5], colors ='k')

ax.set_xlim(xlim)
ax.set_ylim(ylim)

print('KNN Classifier w/ dataset 1:')
plt.show()

#KNN classifier dataset 2

features = ['x1', 'x2']
X = dataset2[features]
X = X.values
y = dataset2.label

kneighs = KNeighborsClassifier(n_neighbors= 41) 
kneighs.fit(X, y)

xlim = (-5, 42)
ylim = (-5, 42)

xx, yy = np.meshgrid(np.linspace(xlim[0], xlim[1], 71), np.linspace(ylim[0], ylim[1], 81))

Z = kneighs.predict_proba(np.c_[xx.ravel(), yy.ravel()])

Z = Z[:, 1].reshape(xx.shape)

fig = plt.figure(figsize= (5, 3.75))
ax = fig.add_subplot(111)

ax.scatter(X[:, 0], X[:, 1], c=y, cmap = 'RdBu')

ax.contour(xx, yy, Z, [0.5], colors ='k')

ax.set_xlim(xlim)
ax.set_ylim(ylim)

print('KNN Classifier w/ dataset 2:')
plt.show()

# Logistic Regression classifier dataset 1

features = ['x1', 'x2']
X = dataset1[features]
X = X.values
y = dataset1.label

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .25, random_state = 0)


logRegr = LogisticRegression()

logRegr.fit(X, y)


xlim = (-5, 42)
ylim = (-5, 42)

xx, yy = np.meshgrid(np.linspace(xlim[0], xlim[1], 71), np.linspace(ylim[0], ylim[1], 81))

Z = logRegr.predict_proba(np.c_[xx.ravel(), yy.ravel()])

Z = Z[:, 1].reshape(xx.shape)

fig = plt.figure(figsize= (5, 3.75))
ax = fig.add_subplot(111)

ax.scatter(X[:, 0], X[:, 1], c=y, cmap = 'RdBu')

ax.contour(xx, yy, Z, [0.5], colors ='k')

ax.set_xlim(xlim)
ax.set_ylim(ylim)

print('Logistic Regression Classifier w/ dataset 1:')
plt.show()


# Logistic Regression classifier dataset 2

features = ['x1', 'x2']
X = dataset2[features]
X = X.values
y = dataset2.label

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .25, random_state = 0)


logRegr = LogisticRegression()

logRegr.fit(X, y)


xlim = (-5, 42)
ylim = (-5, 42)

xx, yy = np.meshgrid(np.linspace(xlim[0], xlim[1], 71), np.linspace(ylim[0], ylim[1], 81))

Z = logRegr.predict_proba(np.c_[xx.ravel(), yy.ravel()])

Z = Z[:, 1].reshape(xx.shape)

fig = plt.figure(figsize= (5, 3.75))
ax = fig.add_subplot(111)

ax.scatter(X[:, 0], X[:, 1], c=y, cmap = 'RdBu')

ax.contour(xx, yy, Z, [0.5], colors ='k')

ax.set_xlim(xlim)
ax.set_ylim(ylim)

print('Logistic Regression Classifier w/ dataset 2:')
plt.show()


```
